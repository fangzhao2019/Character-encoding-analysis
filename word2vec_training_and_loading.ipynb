```python
import warnings#消除警告
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
import logging
import os
from gensim.models import word2vec
from gensim.models import Word2Vec
from tqdm import tqdm
import time
import jieba
jieba.load_userdict('userdict.txt')
```

```python
time1=time.time()

sentSet_cut = []

with open('news.txt', 'r',encoding='utf-8') as f:
    for ele in tqdm(f.readlines()[1:]):
        ele = ele.strip()
        token_sent = list(jieba.cut(ele))
        sentSet_cut.append(' '.join(token_sent))

with open('news_cut.txt', 'w', encoding='utf-8') as f2:
    for sent in sentSet_cut:
        f2.write(sent)
        f2.write('\n')

time2=time.time()
print(u'共用时%d秒'%(time2-time1))
```

```python
# training the word2vec model
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = word2vec.LineSentence('news_cut.txt')
model = Word2Vec(sentences, sg=1, vector_size=300, window=5, min_count=0, negative=3, sample=0.001, hs=1, workers=50)

model.save('news_w2v.model')
model.wv.save_word2vec_format('news_w2v.vector', binary=False)
```
